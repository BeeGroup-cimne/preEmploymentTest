```python
import pandas as pd
```

```python
#Loading the data
file_path = "data_2019.xls"

df = pd.read_excel(file_path, usecols=["d_ini","d_end","kwh"])
df
```

```python
#Clean Null Values
df = df.dropna(axis= 0)
clean_df = df
```

```python
#Set the dates
date_range = pd.date_range(start=df['d_ini'].min(), end=df['d_end'].max(), freq='D')
timeseries_df = pd.DataFrame(index=date_range)
timeseries_df.index.name = 'date'
timeseries_df.head(30)
```

```python
for _, row in clean_df.iterrows():
    #Calculate days of the bill
    date_range = pd.date_range(start=row['d_ini'], end=row['d_end'], freq='D')

    # Calculate mean consumption for the current period
    mean_consumption = row['kwh'] / len(date_range)
    timeseries_df.loc[date_range, 'kwh'] = mean_consumption
```

```python
print(timeseries_df.head(50))
```

```python
# Check for null values in the 'kwh' column
null_values = timeseries_df[timeseries_df['kwh'].isnull()]

# Count null values
print(null_values)
null_count = timeseries_df.isnull().sum().sum()

print("Total Missing Days: " + str(null_count))

#Change Null to NaN
timeseries_df = timeseries_df.fillna("NaN")
```

```python
#Save the TimeSeries data as CSV
timeseries_df.to_csv('timeseries_BeeGroup.csv')
```

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel("data_2019.xls")
pd.set_option('display.max_rows', None)
```

```python
data['d_ini'] = pd.to_datetime(data['d_ini'], format='%d/%m/%Y')
data['d_end'] = pd.to_datetime(data['d_end'], format='%d/%m/%Y')

data.info()
```

```python
data
```

```python
start_date = pd.to_datetime('2019-01-01')
date_range = pd.date_range(start=start_date, end=data['d_end'].max(), freq='D')
```

```python
daily_data = pd.DataFrame({'Date': date_range})
daily_data['kwh'] = np.nan
```

```python
for index, row in data.iterrows():
    date_difference = (row['d_end'] - row['d_ini']).days + 1  # +1 para incluir el último día
    kwh_per_day = row['kwh'] / date_difference
    mask = (daily_data['Date'] >= row['d_ini']) & (daily_data['Date'] <= row['d_end'])
    daily_data.loc[mask, 'kwh'] = kwh_per_day
```

```python
print(daily_data)
```

```python
daily_data.to_csv('daily_data.csv', index=False)
```

```python
import pandas as pd
from datetime import datetime, timedelta
df = pd.read_excel("data_2019.xls")
```

```python
df
```

```python
df = df.iloc[0:10, 0:3]
```

```python
df['d_ini'] = df['d_ini'].astype(str)
df['d_end'] = df['d_end'].astype(str)
```

```python
df.dtypes
```

```python
df
```

```python
year = '2019'
current_date = year + '-01-01'
NAvalue = 'NaN'
result_format = '%d/%m/%Y'
```

```python
result = pd.DataFrame(columns=['date', 'kwh'])

for _, row in df.iterrows():
    # Case where there are missing dates from the current date, to the 
    # Dataframe's ini_date
    if row['d_ini'] != current_date:
        ini_date = current_date
        date_obj = datetime.strptime(row['d_ini'], '%Y-%m-%d')

        day_before = date_obj - timedelta(days=1)

        end_date = day_before.strftime('%Y-%m-%d')
        

        date_range = pd.date_range(start=ini_date, end=end_date)

        temp = pd.DataFrame({'date': date_range})
        temp['kwh'] = NAvalue

        result = pd.concat([result, temp])

    # After solving the previous case if it happened, get d_ini and d_end 
    # from the current row, create the temp dataframe
    # and fill it with the given range and kwh value

    date_range = pd.date_range(start=row['d_ini'], end=row['d_end'])
    temp = pd.DataFrame({'date': date_range})
    temp['kwh'] = row['kwh']

    result = pd.concat([result, temp])

    date_obj = datetime.strptime(row['d_end'], '%Y-%m-%d')
    day_before = date_obj + timedelta(days=1)

    current_date = day_before.strftime('%Y-%m-%d')


result['date'] = result['date'].dt.strftime(result_format)
```
